# File Patch

```python
df.loc[df["EVENT_DESCRIPTION"].str.contains("Train", na = False, regex = False) & df["EVENT_DESCRIPTION"].str.contains("Car", na = False, regex = False)]
```

```python
# First case TrainID
df["TrainID1"] = df["EVENT_DESCRIPTION"].str.extract(r"Train (\d+)") 
df.loc[df["TrainID1"].isna() == False].shape
```

```python
# 2nd case TrainID
df["TrainID2"] = df["ASSET_ID_RAW"].str.extract(r"TR___(\d+)") 
df.loc[df["TrainID2"].isna() == False].shape
```

```python
# 3rd case TrainID
df["TrainID3"] = df["EVENT_DESCRIPTION"].str.extract(r"TR___(\d+)") 
df.loc[df["TrainID3"].isna() == False].shape
```

```python
df.loc[(df["TrainID1"].isna() == False) | 
       (df["TrainID2"].isna() == False) | 
       (df["TrainID3"].isna() == False)].shape
```

```python
# Combine TrainID results
df["TrainID"] = (df["TrainID1"].astype(str) +
                 df["TrainID2"].astype(str) +
                 df["TrainID3"].astype(str)).str.replace("nan", "",regex=None)
df["TrainID"] = df["TrainID"].replace("", np.nan)
del df["TrainID1"], df["TrainID2"], df["TrainID3"]
df.loc[df["TrainID"].isna() == False].shape
```

```python
# First case CarID
df["CarID1"] = df["EVENT_DESCRIPTION"].str.extract(r"Train \d+/Car (\d+)") 
df.loc[df["CarID1"].isna() == False].shape
```

```python
# 2nd case CarID
df[["CarID2", "ServiceID"]] = df["EVENT_DESCRIPTION"].str.extract(r"cars (\d+)/(\d+)") 
df.loc[df["CarID2"].isna() == False].shape
```

```python
# Combine CarID results
df["CarID"] = (df["CarID1"].astype(str) +
                 df["CarID2"].astype(str)).str.replace("nan", "",regex=None)
df["CarID"] = df["CarID"].replace("", np.nan)
del df["CarID1"], df["CarID2"]
df.loc[df["CarID"].isna() == False].shape
```

```python
df.loc[df["ServiceID"].isna() == False].shape
```

```python
df.loc[(df["TrainID"].isna() == False) | 
       (df["CarID"].isna() == False) | 
       (df["ServiceID"].isna() == False)]
```

# File Export

```python
df = df.convert_dtypes()
df.info()
```

```python
df['EQUIPMENT_DATE'] = pd.to_datetime(df['EQUIPMENT_DATE'])
df['ACQUISITION_DATE'] = pd.to_datetime(df['ACQUISITION_DATE'])
df['SCS_TIME'] = pd.to_datetime(df['SCS_TIME'])
df['DATETIME_SENT'] = pd.to_datetime(df['DATETIME_SENT'])
df['DATETIME_RECEIVED'] = pd.to_datetime(df['DATETIME_RECEIVED'])
df['TIME_CODE'] = pd.to_datetime(df['TIME_CODE'])
df.info()
```

```python
df.head()
```

```python
# Sort Data
df = df.sort_values(by='SCS_TIME',ascending=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
# Define File Save Parameters
FileName = "ECS"
Run = "-B0001"
singleSave = True

# Get length of dataframe
df_len = len(df)
# Inspect data
print(df_len)

# Define Size of Partitioned Dataframes
partionSize = 500000

# Define Number of Partitions (Always Round Up to Nearest Interger)
if (df_len == partionSize): 
    partitions = 1
else:
    partitions = df_len // partionSize + 1

# Inspect data
print(partitions)
```

```python
# Export file based on above settings
if singleSave == True:
    df.to_csv(targetFileName, index=False)
    print(targetFileName + " SAVED")
else:
    # Split Dataframe into batches of 500K rows
    for counter in range(partitions):
        startPoint = counter * partionSize
        df_subset = df.iloc[startPoint : (startPoint + partionSize)]
        #print(df_subset.info())

        # Save File
        fileNameN = FileName + Run + '-' + str(counter).zfill(3) + '-alarmsTagged' + '.csv'
        df_subset.to_csv(fileNameN, index=False)
        print(fileNameN + " SAVED")

# Ring Beeper When Complete
import winsound
import time

duration1 = 400  # milliseconds
freq1 = 400  # Hz
duration2 = 600  # milliseconds
freq2 = 300  # Hz
repeatCount = 7
for n in range(repeatCount):
    winsound.Beep(freq1, duration1)
    winsound.Beep(freq2, duration2)
    time.sleep(1)
```
